```python
start_index = 22850

# Define the function to get IMDb ratings
def get_imdb_rating(title, year=None):
    try:
        results = ia.search_movie(title)
        if year:
            results = [m for m in results if 'year' in m and m['year'] == year]
        if not results:
            return None
        movie = ia.get_movie(results[0].movieID)
        return movie.get('rating', None)
    except:
        return None

# List to store new ratings
ratings = []

# Only iterate from index 10833 onward
for index, row in df.iloc[start_index: ].iterrows():
    title = row['title']
    year = row['year'] if 'year' in df.columns else None
    rating = get_imdb_rating(title, year)
    ratings.append(rating)
    print(f"{index+1}/{len(df)}: {title} - {rating}")
    sleep(0.1)  # Rate limiting to avoid hitting IMDb too fast

# Insert ratings back into the original DataFrame (starting at correct row)
df.loc[start_index:, 'rating'] = ratings

# Save to new CSV
df.to_csv('updated_with_ratings.csv', index=False)
```

```python
#Next, let's visualize our data.

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
avg_ratings = df.groupby('release_year')['rating'].mean()
avg_ratings = avg_ratings.sort_index()

plt.figure(figsize=(8, 4))
avg_ratings.plot(kind='bar')

plt.xlabel('Release Year')
plt.ylabel('Average Rating')
plt.title('Average Movie or TV Show Rating by Release Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Plotting release_year and rating shows us there's really no correlation for release_year and how popular a TV show or movie is. So, that debunks whether or not release_year will have an impact on popularity.

```python
#association of tv show genre to rating

tv_df = df.dropna(subset=['listed_in'])
tv_df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
tv_df['genres'] = tv_df['listed_in'].str.split(', ')
exploded = tv_df.explode('genres')
genre_avg_rating = exploded.groupby('genres')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(25, 4))
genre_avg_rating.plot(kind='bar')

plt.title('Average TV Show Rating by Genre')
plt.ylabel('Average Rating')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
